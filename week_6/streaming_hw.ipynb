```python
import numpy as np
import pandas as pd
from datetime import datetime

columns = {
    "lpep_pickup_datetime": str,
    "lpep_dropoff_datetime": str,
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": np.float64,
    "tip_amount": np.float64,
}

df_in = pd.read_csv("../data/green_tripdata_2019-10.csv", usecols=columns.keys(), 
                    dtype=columns, parse_dates=["lpep_pickup_datetime", "lpep_dropoff_datetime"])
# df_out = df_in.head(10)
# df_out.to_csv("../data/green_tripdata_sample.csv", index=False)
df_in.to_csv("../data/green_tripdata_2019-10_filtered.csv", index=False)
```

```python
import csv
import json
import time
from kafka import KafkaProducer


# csv_file = "../data/green_tripdata_2019-10.csv"
TOPIC_NAME = "green-trips"
csv_file = "../data/green_tripdata_2019-10_filtered.csv"


# Create a Kafka producer
producer = KafkaProducer(
    bootstrap_servers="localhost:9092",
    value_serializer=lambda v: json.dumps(v).encode("utf-8")
)

t0 = time.time()

with open(csv_file, "r", newline="", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        # Each row will be a dictionary keyed by the CSV headers
        producer.send(TOPIC_NAME, value=row)

# Make sure any remaining messages are delivered
producer.flush()
producer.close()

t1 = time.time()
print(f'took {(t1 - t0):.2f} seconds')
print(t1-t0)
```
